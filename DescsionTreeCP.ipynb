# Spliting into Categorical and numerical columns

```python
categorical_cols = ["Weather", "Road_Type", "Time_of_Day", "Road_Condition", "Vehicle_Type", "Road_Light_Condition"]
numerical_cols = ["Speed_Limit", "Traffic_Density", "Number_of_Vehicles", "Driver_Alcohol", "Driver_Age", "Driver_Experience"]
```

```python
Transformer = ColumnTransformer([
    ("trfn1", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
], remainder="passthrough")
```

```python
transformed_data = Transformer.fit_transform(df)
```

```python
if not isinstance(transformed_data, np.ndarray):
    transformed_data = transformed_data.toarray()
```

```python
categorical_feature_names = Transformer.named_transformers_["trfn1"].get_feature_names_out(
    input_features=["Weather", "Road_Type", "Time_of_Day", "Road_Condition", "Vehicle_Type", "Road_Light_Condition"]
)
```

```python
numerical_feature_names = df.drop(columns=["Weather", "Road_Type", "Time_of_Day", 
                                           "Road_Condition", "Vehicle_Type", "Road_Light_Condition"]).columns.tolist()
```

```python
all_feature_names = list(categorical_feature_names) + numerical_feature_names
```

```python
print("Transformed Data Shape:", transformed_data.shape) 
print("Feature Names Count:", len(all_feature_names))
```

```python
transformed_df = pd.DataFrame(transformed_data, columns=all_feature_names)
```

```python
transformed_df
```

# Apply Train_Test_Split

```python
X = transformed_df.drop(columns = ["Accident"], axis = 1)
y = transformed_df["Accident"]
```

```python
numerical_cols = ["Speed_Limit", "Traffic_Density", "Number_of_Vehicles", "Driver_Alcohol", "Driver_Age", "Driver_Experience"]
```

```python
scaler = StandardScaler()
```

```python
missing_cols = [col for col in numerical_cols if col not in X.columns]
if missing_cols:
    print(f"Missing columns in DataFrame: {missing_cols}")
else:
    X_scaled = scaler.fit_transform(X[numerical_cols])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.33, random_state=100) 
```

# RandomForestClassifier()

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf_classifier = RandomForestClassifier()

parametrs_fr = {
    'max_depth': [3, 5, 10],
    'n_estimators': [150, 200, 300],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2,4]
}


grid_search_reg_forest3 = GridSearchCV(rf_classifier, parametrs_fr, cv=5)
grid_search_reg_forest3.fit(X_train, y_train)
```

```python
best_params = grid_search_reg_forest3.best_params_
print(best_params)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf_classifier = RandomForestClassifier()

paramtrs_fr = {
    "max_depth": [3],
    "min_samples_leaf": [1],
    "min_samples_split": [2],
    "n_estimators": [150]  
}

grid_search_reg_forest3 = GridSearchCV(rf_classifier, paramtrs_fr, cv=5)
grid_search_reg_forest3.fit(X_train, y_train)

print("Best Train Score:", grid_search_reg_forest3.best_score_)
print("Test Score:", grid_search_reg_forest3.score(X_test, y_test))
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn_classifier = KNeighborsClassifier()

param_knn = {
    "n_neighbors": range(1, 30),  
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "ball_tree", "kd_tree"],
    "leaf_size": range(5, 50)
}

grid_search_knn = GridSearchCV(knn_classifier, param_knn, cv=5)
grid_search_knn.fit(X_train, y_train)
```

```python
best_gs_knn = grid_search_knn.best_estimator_

print('Score on train data = ', round(best_gs_knn.score(X_train, y_train), 4))
print('Score on test data = ', round(best_gs_knn.score(X_test, y_test), 4))
```

# Using Ensemble Voting classifier

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

rf = RandomForestClassifier(**grid_search_reg_forest3.best_params_)
knn = KNeighborsClassifier(n_neighbors=5)

voting_clf = VotingClassifier(estimators=[
    ("rf", rf),
    ("knn", knn),
], voting="hard")  


voting_clf.fit(X_train, y_train)

print("Ensemble Test Score:", voting_clf.score(X_test, y_test))
```

# Using Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier

bagging_clf = BaggingClassifier(
    estimator=RandomForestClassifier(** grid_search_reg_forest3.best_params_),
    n_estimators = 10,
    random_state = 42)

bagging_clf.fit(X_train, y_train)
print("Bagging Test Score:", bagging_clf.score(X_test, y_test))
```

# Using Boosting

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

adaboost_clf = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),  
    n_estimators=50
    learning_rate=0.1,
    random_state=42
)

adaboost_clf.fit(X_train, y_train)

print("AdaBoost Test Score:", adaboost_clf.score(X_test, y_test))
```
